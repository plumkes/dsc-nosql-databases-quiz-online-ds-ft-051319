## Question 1

What is the difference between a NoSQL and a SQL database?  Which is better?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
A SQL database is purely tabular, which works for certain kinds of data but not all. A NoSQL database can store data in different formats with different schemas, or potentially a lack of a solid schema. Some examples are document, key-value, and column stores; graph databases and RDDs. 
There is not one that is better or worse without context; it all depends on the kind of data you are wanting to store.

## Question 3

How does MongoDB work? How is it different from a relational database?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MongoDB uses a Document Store functionality as opposed to just tabular data. In this way you are able to group certain data together in documents that may not be the case for all other data in other documents. The example that MongoDB itself uses is medical information for patients. MongoDB has the capability of storing two documents comfortably next to each other: a patient who has two phone numbers, two addresses, emergency contacts, and 20 prescriptions; and one who has one address, and no prescriptions. 






## Question 4

What is fault tolerance, and what does it have to do with Resilient Distributed Datasets?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
Fault tolerance applies to databases that may have more than one server where data is stored (hence, I believe, where the "reslilient distributed" part of RDD's comes from--resiliant because the data is "fault tolerant," and distributed becuase all your data isn't in one place, or even a few places). Backups to each server will be stored on an alternate server so that if any of the servers go down, you can access the data you need on other servers. 







## Question 5

What is MapReduce? How is it related to Hadoop?

Write your answer below this line: 
_______________________________________________________________________________________________________________________________
MapReduce is a two-step operation that has to do with querying Resilient Distributed Datasets. The "map" part, similar to the **map** function in Python, will execute the same query simultaneously to the different areas where your data is located, and the "reduce" part is referring to aggregating (or possibly calculating) those simultaneous queries that you ran.
